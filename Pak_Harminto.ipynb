```python
import tracemalloc as tc
import pandas as pd
import io
import time as t

# mbukak file dan membaca file
with open("angka_acak_1_20_500.csv",'r') as f:
  filename = f.read()

# fungsi io.String ngge men iso moco filename dadi string, semisal nggak dinei bakal ono tulisan "FILE E KEDAWAN COK"(dalam bahasa halus)
pf = pd.read_csv(io.StringIO(filename))
def bubble(a):
    ijol=0
    arr = list(a)
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
          # arti : jika arr index j luih gede sko arr index j+1 maka pindah ruas yang dimana arr[j], arr[j+1] dadi arr[j+1], arr[j] (7>6 maka sk awal e 7,6 dadi 6,7)
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                ijol+=1
    return arr, ijol
# iki ngge njupuk datane sk iloc i ngge njupuk berdasarkan posisi trus [:,0], : kui  ambil semua baris, 0 kui sko kolom 0/disik dw, nah tolist berarti kui ndadike hasil data kui dadi list [1,4,6,2,7]
data = pf.iloc[:,0].tolist()

# iki nggawe skenarione
# data urut e naik ik
best_case = sorted(data)
# iki acak
average_case = data[:]
# iki urut secara kewalik
worst_case = sorted(data, reverse=True)

# iki ngge menyalakan tracer ram
tc.start()
# iki jalankan bubble sort dalam 4 skenario
_, w_best = bubble(best_case)
_, w_avg = bubble(average_case)
_, w_worst = bubble(worst_case)
_, swap_total = bubble(data[:])


current, peak = tc.get_traced_memory()
# iki mandek e
tc.stop()

# iki kabeh ngge nggolek kompleksitas waktu secara real-time
start = t.time()
bubble(best_case)
end= t.time()
tb = end-start

start = t.time()
bubble(average_case)
end= t.time()
tavg = end-start

start = t.time()
bubble(worst_case)
end= t.time()
tw = end-start

print(" kompeksitas ruang ".center(50,'-'))
print(f"Penggunaan memori saat ini: {current/1024:.2f} KB")
print(f"Puncak penggunaan memori : {peak/1024:.2f} KB\n")

print("Pertukaran swap(Total,Best,Average,Worst)".center(80,'-'))
print(f'Pertukaran swap : {swap_total}')
print(f"Jumlah pertukaran (Best Case)   : {w_best}")
print(f"Jumlah pertukaran (Average Case): {w_avg}")
print(f"Jumlah pertukaran (Worst Case)  : {w_worst}\n")

print("Kompleksitas waktu real-time(Best,Average,Worst)".center(80,'-'))
print(f"Jumlah waktu real-time (Best Case)   : {tb:.6f} detik")
print(f"Jumlah waktu real-time (Average Case): {tavg:.6f} detik")
print(f"Jumlah waktu real-time (Worst Case)  : {tw:.6f} detik")

# iki ngge menunjukan sedikit isine
pf.head()


# Catatan kui ijol penting jebul, tiwas mau dihapus, bajilak i python iso protes jebul,,, kui sk ijoi ki ngge pertukaran e
```
