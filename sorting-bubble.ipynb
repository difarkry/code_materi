```
import tracemalloc as tc
import pandas as pd
from google.colab import files

upload = files.upload()
filename = list(upload.keys())[0]

pf = pd.read_csv(filename)

def bubble(a):
    global arr
    w = 0
    arr = list(a)
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                w += 1
    return w

data_csv = pf.iloc[:,0].tolist()

best_case = sorted(data_csv)
average_case = data_csv[:]
worst_case = sorted(data_csv, reverse=True)

tc.start()
w_best = bubble(best_case)
w_avg = bubble(average_case)
w_worst = bubble(worst_case)
current, peak = tc.get_traced_memory()
tc.stop()

print(f"Penggunaan memori saat ini: {current/1024:.2f} KB")
print(f"Puncak penggunaan memori : {peak/1024:.2f} KB")
print(f'Kompleksitas waktu : {bubble(data_csv)}')
print(f"Jumlah pertukaran (Best Case)   : {w_best}")
print(f"Jumlah pertukaran (Average Case): {w_avg}")
print(f"Jumlah pertukaran (Worst Case)  : {w_worst}")
print(arr)
print(pf)
```
